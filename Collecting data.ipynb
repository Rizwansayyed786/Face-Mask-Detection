```python
while True:
    cv2.imshow('Result',image)
    if cv2.waitKey(2) == 27:
        break 
cv2.destroyAllWindows()
```

```python
df=cv2.CascadeClassifier('haar_data.xml')
```

```python
df.detectMultiScale(image)
```

```python
while True:
    faces=df.detectMultiScale(image)
    for x,y,w,h in faces:
        cv2.rectangle(image,(x,y),(x+w,y+h),(255,255,255),4)
    cv2.imshow('Result',image)
    if cv2.waitKey(2) == 27:
        break 
cv2.destroyAllWindows()
```

```python
# capturing images from default camera

capture = cv2.VideoCapture(0) # 0 for default camera 1 for another installed camera
data=[]
while True:
    
        flag,image = capture.read() # if flag is working its gonn detect the face
        
        if flag: 
            faces = df.detectMultiScale(image)
            
            for x,y,w,h in faces:
                cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,255),4)
                face=image[y:y+h,x:x+w,:]
                face=cv2.resize(face,(50,50))
                print(len(data))
                if len(data) < 400:
                    data.append(face)
                
            cv2.imshow('Result',image)
                # 27 - ASCII of escape
                
            if cv2.waitKey(2) == 27 or len(data) >= 200:
                    break 
                    
capture.release()
cv2.destroyAllWindows()
```

```python
np.save('with Mask.npy',data)
```

```python
np.save('without Mask.npy',data)
```
